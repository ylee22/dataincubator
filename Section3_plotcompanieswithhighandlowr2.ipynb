```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# plot change in # of employees to stock price change
def plotchange(sdiff, ediff):
    if ediff.max() > sdiff.max():
        scaling = ediff.max()/sdiff.max()
        scaledediff = ediff/scaling
    else:
        scaledediff = ediff
    plt.plot(sdiff)
    plt.plot(scaledediff)
    plt.xticks(rotation='vertical')
    plt.show()
```

```python
# read the pre processed data
df = pd.read_csv('deltastemr2.csv')
```

```python
df = df.sort_values(by=['r2'], ascending=False)
c = list(df.names)
```

# Top 6 companies (Itron, Amber Road, INTL FCStone, TESSCO Technologies, Trans World Entertainment, Tandem Diabetes Care) with highest correlation between monthly stock price change and the change in the number of employees

```python
# plot scaled employee difference (scaled to visualize both) to the stock change
temp = pd.read_csv('deltastem/' + c[1] + '.csv')
temp = temp.set_index('as_of_date')
# plot the differences in stock value to employee number fluxuations
plotchange(temp.close, temp.employees_on_platform)
```
